```python
#import dependencies
import pandas as pd
import sqlalchemy as db
import matplotlib.pyplot as plt
import psycopg2
from config import pword
from config import username
engine_url = 'postgresql+psycopg2://'+username+':'+pword+'@localhost/EmployeeSQL'
engine = db.create_engine(engine_url)
connection = engine.connect()
metadata = db.MetaData()
```

```python
#create histogram of salaries
salaries = db.Table('salaries', metadata, autoload=True, autoload_with=engine)
query = db.select([salaries.columns.salary])
ResultProxy = connection.execute(query)
ResultSet = ResultProxy.fetchall()
print(ResultSet)
```
